```python
def lengthOfLongestSubstring( s: str) -> int:
    #simple implementation 
    #O(n^2) complexity 
    #bit shitty could implement merge algorithm for better complexity
    #
    #
    
    substrings_len = [0]
    for i, letter in enumerate(s):
        sub_substring_len  = 1
        sub_substring = [letter]
        j = i + 1 
        while (j < len(s)) and (s[j] not in sub_substring):
            sub_substring.append(s[j])
            sub_substring_len += 1
            j += 1
        substrings_len.append(sub_substring_len)
    substrings_len.sort()
    return substrings_len[-1]
lengthOfLongestSubstring("")
```
