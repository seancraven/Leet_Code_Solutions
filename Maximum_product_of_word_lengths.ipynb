```python
class Solution:
    def maxProduct(self, words) -> int:
        #make tuple (word,len) # O(n)
        #sort tuple #O(n)
        #Take Largest search for no overlap, recurse untill found #O(n^2)
        #Note Comlpexity of this function is not actually O(n^2) but could 
        #be with set abcedefhi...yz giving constant time look up for does word contain these char
        #Then function is O(n^2)
        tuple_list = [(len(word),word) for word in words]
        tuple_list.sort()
        ans = 0
        max_ans = 0 
        while tuple_list:
            (c_len, c_word) = tuple_list.pop(-1)
            for (length, word) in reversed(tuple_list):
                if not string_overlap(word,c_word):
                    ans = length*c_len
                    if ans > max_ans: max_ans = ans   
        return max_ans
        
def string_overlap(word_1,word_2): #O(w_1 + w_2) => this could be made faster by noting max set length 
    s_w1 = set(word_1)             # has a maximum length of 26 giving you constant time but really
    s_w2 = set(word_2)             # solution is fine
    if len(s_w1.union(s_w2)) < len(s_w1) + len(s_w2):
        return True
    else: return False

Solution.maxProduct( _,["a","ab","abc","d","cd","bcd","abcd"])
```
