```python
import cProfile
class Solution:
    def smallestStringWithSwaps( s: str, pairs) -> str:
        ''' Finds the lexographically smallest string for shortest string problem'''
        free_index = get_linked_pairs(pairs)
        sorted_list = free_index_sort(free_index,s)
        return sorted_list  
```

```python
def asc2_list_to_string(asc2_list: list):
    string = ""
    for i in asc2_list:
        string += chr(i)
    return string

def free_index_sort(free_index_list,string)->list:
    list_to_sort = [letter for letter in string]
    for ind_list in free_index_list: 
        sorted_subset = [list_to_sort[i] for i in ind_list]
        sorted_subset.sort()
        for i,entry in zip(ind_list,sorted_subset):
            list_to_sort[i] = entry
    return ''.join(list_to_sort)

def get_linked_pairs(pairs_list):
    dupe_list = []
    for pair in pairs_list:
        if pair[0] == pair[1]:
            dupe_list.append(pair)
    for duplicate in dupe_list:
        pairs_list.remove(duplicate)
    while overlap(pairs_list): # Loops over sublists till no overlap
        to_remove = []
        group_0 = pairs_list.pop(0)
        for sub_list in pairs_list:
            if overlap([group_0,sub_list]):
                group_0  = unique_append(group_0,sub_list)
                to_remove.append(sub_list)
        pairs_list.append(group_0)
        for entry in to_remove:
            pairs_list.remove(entry)
    return pairs_list
        

def overlap(indecies):
    flat_list = [item for sublist in indecies for item in sublist]
    unique_len = len(list(set(flat_list)))
    return (unique_len != len(flat_list))
        

def unique_append(list_1,list_2):
    list_1.extend(list_2)
    return list(set(list_1)) 

```
