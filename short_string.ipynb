```python
def free_index_sort(free_index_list: list,list_to_sort:list)->list:
    for list in free_index_list: 
        sorted_subset = [list_to_sort[i] for i in list]
        sorted_subset.sort()
        for i,entry in zip(list,sorted_subset):
            list_to_sort[i] = entry
    return list_to_sort
```

```python
from collections import defaultdict
class UnionFind:
    def __init__(self, n):
        self.size = n
        self.parent = [i for i in range(n)]
        
    def find(self, u):
        if u != self.parent[u]:
            self.parent[u] = self.find(self.parent[u])  # Path compression
        return self.parent[u]
    
    def union(self, u, v):
        pu = self.find(u)
        pv = self.find(v)
        if pu == pv: return False
        self.parent[pu] = pv
        return True
    
    def getGroups(self):
        groups = defaultdict(list)
        for u in range(self.size):
            groups[self.find(u)].append(u)
        return groups.values()
        

class Solution:
    def smallestStringWithSwaps( s: str, pairs ) -> str:
        n, m = len(s), len(pairs)
        uf = UnionFind(n)
        for u, v in pairs:
            uf.union(u, v)
            
        groups = uf.getGroups()
        print(groups)
        res = ['-'] * n
        for group in groups:
            chars = [s[i] for i in group]
            chars.sort()
            group.sort()
            for i, c in zip(group, chars):
                res[i] = c
        return "".join(res)
```
