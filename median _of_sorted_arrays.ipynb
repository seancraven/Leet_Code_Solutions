```python
import numpy as np 
import math 
```

```python
class Solution:
    def findMedianSortedArrays(self, nums1, nums2) -> float:
        #build set of both is O(m+n)
        #have to form a tree of some sort
        #Recursively split and follow path to middle vertex
        nums1 = Node(nums1)
        nums2 = Node(nums2)
        mp1, mp2 = median_position(nums1, nums2)
        if mp2:
            sol_2 = find_array_med(nums1, nums2, mp2)
            sol_1 = find_array_med(nums1, nums2, mp1)
            return (sol_2 + sol_1)/2
        else:
            sol = find_array_med(nums1, nums2, mp1)
            return(sol)




def median_position(a1, a2):
    mp2 = None 
    if (len(a1)+ len(a2))%2 == 0:
        mp1 = (len(a1) + len(a2))//2
        mp2 = (len(a1) + len(a2))//2 +1
    else:
        mp1 = math.ceil(((len(a1) + len(a2))/2))
    return mp1, mp2


def find_array_med(a1, a2, mp1): #given that l1.max <= r1.max
    #in dev dont worry about mp2
    if a1.max < a2.min:
        if mp1 <= len(a1):
            median = a1[mp1-1]
        else:
            mp1 = mp1- len(a1)
            median = a2[mp1-1]
        return median
    elif a2.max < a1.min:
        if mp1 <= len(a2):
            median = a2[mp1-1]
        else:
            mp1 = mp1- len(a2)
            median = a1[mp1-1]
        return median
    elif a1.max == a1.min == a2.max == a2.min:
        return a1.max
    else:
        #split into three recursively, loosing one or finding median each time. 
        if (len(a1) > 1):
            l1, r1 = a1.split()
            if (l1.max >= a2.max) and (len(l1) + len(a2) <= mp1):
                return find_array_med(a2,l1, mp1)
            elif (a2.min >= l1.max):
                if len(l1)+ len(a2) == mp1: return a2[-1]
                else:
                    mp1 -= len(l1)
                    return find_array_med(a2, r1,mp1)
            else:
                l2, r2 = a2.split()
                if len(l1) + len(l2) <= mp1:
                    return find_array_med(l1,l2,mp1)
                else:
                    mp1 -= len(l1) + len(l2)
                    return find_array_med(r1,r2,mp1)
        else:
            return find_array_med(a2,a1,mp1)
    
    


        #If len(m) < len(n): and m[[0,-1]] inside n[[o,-1]]
class Node: #node of sorted array split in binary split tree
    def __init__(self,array):
        self.array = array
        if self.array:
            self.min = self.array[0]
            self.max = self.array[-1]
        else:
            self.min = -10**6+1
            self.max = -10**6+1
    def __len__(self):
        return len(self.array)
    def __getitem__(self, key):
        return self.array[key]
    
    def split(self):
        if len(self.array) == 1:
            return Node(self.array) , None

        c = len(self.array)//2
        left = Node(self.array[:c])

        right = Node(self.array[c:])
        return left , right

```
