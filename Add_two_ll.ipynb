```python
import math
# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next


class Solution:
    def addTwoNumbers(self, l1, l2):
        sum = 0
        i = 0 
        while l1:
            sum += l1.val*10**i
            l1 = l1.next
            i += 1
        i = 0
        while l2:
            sum += l2.val*10**i
            l2 = l2.next
            i += 1
        i = 0
        st = str(sum)
        ls = [int(i) for i in st]
        output = None
        for i in ls:
            output = ListNode(i,next = output)
        return output
```
